```python
import gradio as gr
from skimage.transform import resize
import tensorflow
import numpy as np
from tensorflow.keras.models import load_model
import matplotlib.pyplot as plt
```

```python
model = load_model('/content/drive/MyDrive/Colab_Notebooks/wildanimals/wildanimals.h5')
```

```python
labels = {0:'cheetahs', 1:'elephants', 2:'lions', 3:'tigers'}
```

```python
def classify_image(inp):
  inp = resize(inp, (224, 224))
  inp = inp.reshape((1, 224, 224, 3))
  predictions = model.predict(inp)
  return(np.argmax(predictions))
```

```python
img = '/content/drive/MyDrive/Colab_Notebooks/wildanimals/lion1.jpg'
img = plt.imread(img)
preds = classify_image(img)
print( labels[preds])
```

```python
image = gr.inputs.Image(shape=(224, 224))
label = gr.outputs.Label(num_top_classes=3)

gr.Interface(fn=classify_image, inputs=image, outputs=label, capture_session=True).launch(debug=True)
```
